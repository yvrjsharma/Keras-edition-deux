In the real world we don't start with the data rather we start with a problem. We can adopt a universal wworkflow to handle a problem using machine learning approaches. It is structured in three parts -
1. Define the task
2. Develop a model
3. Deploy the model

## Defining the task

Develop a deep understanding of context of the problem

**Frame the problem**

* Detailed discussions with stakeholders
* Identify what will be the input data
* What are we trying to predict?
* You might have to gather new data and annotate it as well in this stage.
* What type of ML problem is it? Regression, binary/multiclass classification, multilabel classification, image segmentation, ranking, clustering, text/image generation, reinforcement learning, and so on.
* Identify what existing solution to the problem looks like.
* Are there any system or business constraints you have to work within?

As result of this exercise, you will be able to produce a *Hypothesis* for the problem. Clearly state what the inputs and targets are, and what is the nature/type of ML problem you are dealing with here.


**Collect a dataset**

* In previous chapter we learned that a model's generalizing power comes from how good the input data is -- large number of data points, reliable labels, and good quality features.
* Once you have the data collected, you might have to figure ot way to annotate it as well.
* Keep in mind that the ata used in traininig is a representative to actal production data.
* **Concept drift** occurs when the properties of production data change overtime, causing model acuracy to decay. This can be countered by constant data collection, clear annotation and timely model traiining. 

**Understand you data**

* Explore and visualize your data to gain insights abbout what features are important, can you do feature engineering, do you have to do feature selection as well and so on.
* Identify patterns in your data, gather information about range of numerical features and frequency of categorical features.
* Deal with missing data.
* What are the toatal samples for each class in a classification problem?
* **Target Leaking** - Presence of features in your data that provide information about the targets and which might not be available in production.

**Choose the right measure of success**

* You can only control something if you can first measure/observe it.
* First you have to define what you would mean by succes. Is it Accuracy? Or is iit Precision and Rcalll? Root Mean Square Error? This should align with high level business goals and succcess.
* Example for Binary balanced class problems - the Accuracy or the area under the Receiver Operating Characteristic (ROC) curve, also called ROC AUC, is used. Example for imbalanced and multi-class problems, the weighted accuracy or weighted ROC AUC, or mainly the Precision and Recall are used.



## Develop a model

Once you are able to iidentify measure of success and have gathered enough data you can start model evelopment.

**Prepare the data**

* Make the raw-data at hand more amenable for deep learninig models.
  
  * Vectorization
    * All inputs and targets in a neeural network must be tensors of floating points.
    * First turn your data into tensors, this is called *Data Vectorization*

  * Normalization
    * Normalize your integer data such that a feature has standard deviation as 1 and mean as 0.
    * Large integer data or high range heterogenous data should be normalized otherwise it might trigger large gradient updates that might destabilize the network and moreobver prevent the network from converging.
    * Ideally all values should be between 0-1. And all features should have homogenous set of values in same range.

  * Handling missing data
    * Either discard the feature alltogether if there are lots of samples with missing values.
    * Or, if it i a categorical feature, create a new category for missing values. If it is a numerical field, replace with median or mean or you can even predict these values using a model with rest of the features as independent features and this field as target.

**Choose an Evalluation protocol**

* As covered in previous chapter/post, you can choose from three common evaluation protocols - A hold out validation set, K-Fold validation, and lastly Iterated shuffled k-fold validation.
* Make sure your validation set is representative of the production data.
* Also kae sure there are no redundatnt samples between training set and validation set.
